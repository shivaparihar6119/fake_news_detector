```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/fake_or_real_news.csv")
```

```python
df['result']=df['label'].apply(lambda x:0 if x=="FAKE" else 1)
df=df.drop("label",axis=1)
df
```

```python
print(df['result'].value_counts())
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
ps=PorterStemmer()
corp=[]
for i in range(0,len(df)):
    ntxt = re.sub('[^a-zA-Z]', ' ', df['text'][i])
    ntxt = ntxt.lower()
    ntxt = ntxt.split()
    ntxt = [ps.stem(word) for word in ntxt if not word in stopwords.words('english')]
    ntxt=' '.join(ntxt)
    corp.append(ntxt)
```

```python
corp[3]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer(max_features=5000,ngram_range=(1,3))
x=vect.fit_transform(corp).toarray()
```

```python
y=df['result']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
from sklearn.svm import LinearSVC
svc_model=LinearSVC()
svc_model.fit(x_train,y_train)
score=svc_model.score(x_test,y_test)
print(score)
y_pred=svc_model.predict(x_test)
cmatrix=confusion_matrix(y_test,y_pred)
print(cmatrix)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb_model=MultinomialNB()
nb_model.fit(x_train,y_train)
score=nb_model.score(x_test,y_test)
print(score)
y_pred=nb_model.predict(x_test)
cmatrix=confusion_matrix(y_test,y_pred)
print(cmatrix)
```
